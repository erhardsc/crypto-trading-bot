```python
import config
import pandas as pd
import sys
import plotly.offline as py
import plotly.graph_objs as go
%matplotlib inline
%matplotlib notebook
py.init_notebook_mode(connected=True)
```

```python
path = config.CONSTS["PATH"]
indicators = pd.read_csv(path + "/data/indicators.csv")
indicators = indicators[['date','momentum','movingAverage']]
candlesticks = pd.read_csv(path + "/data/candlesticks.csv")
candlesticks = candlesticks[['date','open','high','low','close','volume','weightedAverage',]]
merged = indicators.merge(candlesticks, on='date', left_index=False, right_index=False, how="inner")
merged.to_csv(path + '/data/merged_data.csv', index=False)
```

```python
data = pd.read_csv(path + '/data/merged_data.csv')
```

```python
price_trace = go.Scatter(x=data['date'], y=data['weightedAverage'], name= 'Price')
momentum_trace = go.Scatter(x=data['date'], y=data['momentum'], name= 'Momentum')
momentum_trace = go.Scatter(x=data['date'], y=data['movingAverage'], name= 'Moving Average')

plot_data = [price_trace, momentum_trace]
py.iplot(plot_data)
```
